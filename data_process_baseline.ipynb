```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split,StratifiedKFold,train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error,roc_auc_score
from sklearn import linear_model
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
oline = pd.read_excel('oline.xlsx')
oline.head()
```

```python
oline=oline.drop('id',axis=1,inplace=False)
oline=oline.drop('时间',axis=1,inplace=False)
oline
```

```python
oline1=oline.apply(lambda x : (x-np.min(x))/(np.max(x)-np.min(x)))
df=oline1.apply(lambda x:x.var())#方差分析
df=df.sort_values()
df=df.tail(90)
df=df.index
df
```

```python
oline2=oline1.drop('RON_LOSS',axis=1,inplace=False)#筛选互相之间相关性最小的90个变量
df1=oline2.corr()
x={}
y={}
variables=[]
for index in df1.index:
    df1_1=df1.loc[index]
    min=abs(df1_1).min()
    for column in df1.columns:
        if abs(df1_1[column])==min:
            if column in x.keys():
                if min<x[column]:
                    x[column]=min
            else:
                x[column]=min
            y[index]=min
x_=sorted(x.items(),key = lambda x:x[1],reverse = True)
y_=sorted(y.items(),key = lambda x:x[1],reverse = True)
for i in range(90):
    variables.append(x_[i][0])
    variables.append(y_[i][0])
set_=set(variables)
set_
```

```python
oline3=oline
oline3['RON_LOSS']=oline3['RON_LOSS']/oline3['材料辛烷值RON']#选择和因变量相关性最大的90个变量
df2=oline3.corr()
df2_=df2['RON_LOSS']
z=[]
for i in df2_.values:
    z.append(abs(i))
z=np.array(z)
ser=pd.Series(z,index=df2_.index)
ser=ser.sort_values(ascending=False)
ser_=ser[0:91]
index_=ser_.index
index_
```

```python
X=oline.drop('RON_LOSS',axis=1,inplace=False)#F回归选择变量
y=oline['RON_LOSS']/oline['材料辛烷值RON']
X_new=SelectKBest(f_regression,k=90).fit_transform(X,y)
X_new
```

```python
X_new_=X_new.T
```

```python
new_X_index=[]
for i in range(90):
    for column in X.columns:
        if [j for j in X_new_[i]]==[j for j in np.array(X[column])]:
            new_X_index.append(column)
new_X_index
```

```python
list1=[new for new in df if new in set_]#取交集
list2=[new for new in list1 if new in index_]
list3=[new for new in list2 if new in new_X_index]
list3.append('材料辛烷值RON')
list3
```

```python
data1=data.apply(lambda x:(x - np.mean(x)) / np.std(x))#z_score标准化
data1
```

```python
alphas_to_test=np.linspace(0.001,1)#岭回归
X_train, X_test, y_train, y_test = train_test_split(data1, label, test_size=0.3, random_state=42)
model=linear_model.RidgeCV(alphas=alphas_to_test,store_cv_values=True)
model.fit(X_train,y_train)
y_preditc=model.predict(X_test)
mse_test=np.sum((y_preditc-y_test)**2)/len(y_test)
rmse_test=mse_test ** 0.5
rmse_test
```
